```python
import glob    
import os
import numpy as np
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
trading212_tickers = list(np.load('trading212tickers.npy'))
```

```python
earnings_dir = f'earnings/'
all_items = os.listdir(earnings_dir)
dates = [item for item in all_items if os.path.isdir(os.path.join(earnings_dir, item))]
dates_datetime = sorted([datetime.datetime.strptime(date, '%Y-%m-%d')  for date in dates])
```

```python
df = {'Date': [], 'Ticker': [], 'Start': [], 'End': [], 'TV': [], 'Zacks': []}

k = 0
for  date in dates_datetime:

    year, month, day = date.year, date.month, date.day
    print('############################')
    print(f'## {year} / {month} / {day}')
    print('############################')

    earnings_date = str(datetime.date(year, month, day))
    start_date = str(datetime.date(year, month, day) - datetime.timedelta(days = 7))
    end_date = str(datetime.date(year, month, day) + datetime.timedelta(days = 7))
    directory = f'earnings/{year}-{month}-{day}/'
    all_items = os.listdir(directory)
    companies = [item for item in all_items if (os.path.isdir(os.path.join(directory, item)) and item in trading212_tickers)]
    for company in companies:
        ticker = yf.Ticker(company)
        volumean = ticker.history()['Volume'].mean()
        company_dir = directory + company + '/'
        try:
            with open(company_dir + 'tradingview.com', 'r') as f:
                rank_tradingview = f.read().split(' ')[-1]
            with open(company_dir + 'zacks.com', 'r') as f:
                rank_zacks = f.read().split(' ')[-1]
                rank_tradingview = float(rank_tradingview)
                rank_zacks = float(rank_zacks)
        
            print('\n')
            print(company)
            print(round(rank_tradingview,2), rank_zacks)


            yf_ticker = yf.download(tickers=company, start=start_date, end=end_date, progress=False, ignore_tz=True, interval = '1h', prepost = True)
            values = yf_ticker['Open'].iloc[:].values
            earnings_indices = yf_ticker.index.get_loc(earnings_date)
            # plt.figure(figsize = (20,4))
            # plt.axvspan(earnings_indices.start, earnings_indices.stop, color = 'red', alpha = 0.5)
            # plt.plot(values, marker = 'o', color = 'k')
            # plt.show()
            # plt.close()


            df['Date'].append(f'{year}-{month}-{day}')
            df['Ticker'].append(company)
            df['Start'].append(values[0])
            df['End'].append(values[-1])
            df['TV'].append(round(rank_tradingview,2))
            df['Zacks'].append(rank_zacks)

        except:
            continue
    

        
```

```python
pdf = pd.DataFrame(df)
```
